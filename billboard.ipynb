```python
import billboard
import pandas as pd
import json
import datetime
```

```python
#create date range 
from datetime import timedelta, date
dateList=[]
def daterange(date1, date2):
    for n in range(int ((date2 - date1).days)+1):
        yield date1 + timedelta(n)

start_dt = date(2017, 1, 1)
end_dt = date(2017, 2, 1)
for dt in daterange(start_dt, end_dt):
    dateList.append(dt.strftime("%Y-%m-%d"))
print(dateList)
```

```python
#print out sample response
response = billboard.ChartData('hot-100').json()
print(response)
```

```python
#Loop through dates and retrieve info
chart_data = []

for date in dateList:
    response = billboard.ChartData('hot-100', date=date).json()
    response_object = json.loads(response)
    chart_data.append(response_object)
   
```

```python
#print(chart_data)
print(json.dumps(chart_data, indent=4, sort_keys=False))
```
