```python
#import dependencies
import pandas as pd
import json
import datetime
import csv
import requests
import os
import spotipy

#import api key
#from config import api_key
from last_fm_config import api_key, secret_key
from config import client_id, client_secret
```

```python
billboard_csv = os.path.join("billboard_lyrics_1964-2015.csv")
```

```python
rank = []
song = []
no_featuring = []
year = []
lyrics = []
source = []


with open(billboard_csv, newline="") as csvfile:
    csvreader=csv.reader(csvfile, delimiter=",")
    next(csvreader, None)
    for row in csvreader:
        rank.append(row[0])
        song.append(row[1])
        new_artist = row[2].split("featuring")
        no_featuring.append(new_artist[0])
        year.append(row[3])
        lyrics.append(row[4])
        source.append(row[5])
cleaned_csv = zip(rank,song,no_featuring,year,lyrics,source)
```

```python
output_file = os.path.join("billboard_lyrics_no_featuring.csv")

with open(output_file,'w', newline="") as datafile:
    csvwriter = csv.writer(datafile, delimiter=',')
    csvwriter.writerow(['Rank','Song','Artist','Year','Lyrics','Source'])
    csvwriter.writerows(cleaned_csv)
```

```python
tracks_per_year = billboard_df.loc[billboard_df["Year"]==2015]
tracks_per_year["Album"] = ""
tracks_per_year.head()
```

```python
info = []
album_list = []
for index, row in tracks_per_year.iterrows():
    base_url = "http://ws.audioscrobbler.com/2.0/?method=track.getInfo&api_key="
    
    artist = row["Artist"]
    song = row["Song"]
    
    track_info = requests.get(f"{base_url}{api_key}&artist={artist}&track={song}&format=json")
    track_info = track_info.json()
    info.append(track_info)


print(json.dumps(info, indent=4, sort_keys=True))
```

```python

for track in info:
    try:
        album_list.append(track["track"]["album"]["title"])
    except (KeyError):
        album_list.append("N/A")

  
print(album_list)
#tracks_per_year
```
